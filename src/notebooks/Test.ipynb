```python
import os
import polars as pl
from pathlib import Path

from src.modules.preprocessing import text_preprocessing

def preprocess_s140_dataset(raw_data_path, preprocessed_data_path):
    # Load the dataset
    encode = 'ISO-8859-1'
    df = pl.read_csv("../../data/training.1600000.processed.noemoticon.csv", has_header=False, encoding=encode, new_columns=["polarity", "id", "date", "flag", "user", "text"])
    
    # Create a new column for the preprocessed text
    df = df.with_column(
        df["text"].apply(lambda text: preprocessing.text_preprocessing(text)).alias("preprocessed_text")
    )

    # Save the preprocessed dataset
    df.write_csv(preprocessed_data_path)


def preprocess_f1_dataset(raw_data_path, preprocessed_f1_data_path):
    # Load the dataset
    dtypes = {
        'user_name':pl.Utf8,
        'user_location':pl.Utf8,
        'user_description':pl.Utf8,
        'user_created':pl.Utf8,
        'user_followers':pl.Utf8,
        'user_friends':pl.Utf8,
        'user_favourites':pl.Utf8,
        'user_verified':pl.Utf8,
        'date':pl.Utf8,
        'text':pl.Utf8,
        'hashtags':pl.Utf8,
        'source':pl.Utf8,
        'is_retweet':pl.Utf8
    }
    encode = 'Utf8'
    df = pl.read_csv("../data/F1_tweets.csv", dtypes=dtypes, encoding=encode)
    
    # Create a new column for the preprocessed text
    df = df.with_column(
        df["text"].apply(lambda text: preprocessing.text_preprocessing(text)).alias("preprocessed_text")
    )

    # Save the preprocessed dataset
    df.write_csv(preprocessed_f1_data_path)
```
